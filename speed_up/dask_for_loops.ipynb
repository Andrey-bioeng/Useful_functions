```python
from dask.delayed import delayed

from dask.distributed import Client, progress
client = Client(threads_per_worker=4, n_workers=4)
client
```

```python
list_ = [1,4,1,50]

@delayed
def square(a):
    return a**2

@delayed
def sum_list(l):
    return sum(l)
    
square = delayed(square)
sum_list = delayed(sum_list)

step1 = [square(a) for a in list_]
step2 = sum_list(step1)
step2.visualize()
```

```python
step3 = step2.compute()
```
